#Parameter Fungsi Print

```python
Tanggal = 28
Bulan = "Februari"
Tahun = 2010

#28-Februari-2010
print(Tanggal)
print(Bulan)
print(Tahun)
```

```python
#28-Februari-2010
print(Tanggal, Bulan, Tahun, sep="-")
```

```python
#28-Februari-2010
print(Tanggal, end="-")
print(Bulan, end="-")
print(Tahun)
```
