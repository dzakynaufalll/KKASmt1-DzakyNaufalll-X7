#Aritmatika

```python
#Tanpa di simpan ke variabel
print(5000*3)
```

```python
#Simpan operand ke variabel

harga_cilok=6000
banyak_porsi=3

print(harga_cilok*banyak_porsi)
```

```python
#Simpan hasil operasi ke variabel

jumlah_uang_yang_dibayar = 7000*3

print(jumlah_uang_yang_dibayar)
```

```python
harga_kojek=10000
jumlah_porsi_kojek=5
total_harga_kojek=harga_kojek*jumlah_porsi_kojek

print(total_harga_kojek)
```

```python
uang_yang_diberi =100000
uang_kembalian= uang_yang_diberi- total_harga_kojek

print(uang_kembalian)
```

```python
# 1. Hitung langsung di print
print(24*3)
```

```python
# 2. Simpan hasil perkalian ke variabel total_jumlah_botol

total_jumlah_botol = 24 * 3
print(total_jumlah_botol)
```

```python
# 3. Simpan operand ke variabel lalu hitung langsung di print()
banyak_botol_per_kardus = 24
jumlah_kardus = 3
print(banyak_botol_per_kardus * jumlah_kardus)
```

```python
# 4. Simpan operand ke variabel, lalu hasil perkaliannya juga disimpan ke variabel lain
banyak_botol_per_kardus = 24
jumlah_kardus = 3
total_jumlah_botol = banyak_botol_per_kardus * jumlah_kardus
print(total_jumlah_botol)
```

#Aplikasi Menghitung Luas Persegi Panjang

```python
panjang = int(input("Masukkan panjang persegi panjang (cm): "))
lebar = int(input(" Masukkan lebar persegi panjang (cm): "))

luas = panjang*lebar
print(f"luas persegi panjang yaitu {luas} cm^2.")
```

```python
panjang = int(input("Masukkan panjang balok (cm): "))
lebar = int(input(" Masukkan lebar balok (cm): "))
tinggi = int(input(" Masukkan tinggi balok (cm): "))
volume = panjang*lebar*tinggi
print(f"volume balok yaitu {volume} cm^3.")
```
